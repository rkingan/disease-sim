# Example: Getting eigenvalues and eigenvectors from the adjacency matrix of a graph

```python
from dissim.centrality import get_numpy_adj_matrix, largest_eigenvalue
from dissim.igraph_util import load_graph_here
import numpy as np
import igraph
```

```python
# Load the karate graph and plot it
KG = load_graph_here("karate")
igraph.plot(KG, bbox=(250, 250))
```

```python
adj = get_numpy_adj_matrix(KG)
print("Type of matrix: ", type(adj))
print("Shape of matrix:", adj.shape)
```

```python
w, v = np.linalg.eig(adj)
```

```python
print("Shape of eigenvalues:", w.shape)
print("Shape of eigenvectors:", v.shape)
```

```python
print(w)
```

```python
w_real = np.real(w)
print(w_real)
```
